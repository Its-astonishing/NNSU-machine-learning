## Разделим данные

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('price_range',axis=1)
y=data['price_range']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
def f(n):
    for i in range(1, n + 1):
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(X_train, y_train)
        print('Ошибка kNN с n_neighbors =', i ,'составляет', (1 - knn.score(X_test, y_test)) * 100, '%')
f(15)        
```

Оптимально взять k=9

Вывод: Получаем ошибку меньше, чем с K=9. Теперь ошибка составляет около 6.4%

Посчитаем характеристики предсказания и построим матрицу рассеивания

```python
knn9 = KNeighborsClassifier(n_neighbors=9)
knn9.fit(X_train, y_train)
from sklearn.metrics import classification_report, confusion_matrix
pred = knn9.predict(X_test)
print(classification_report(y_test, pred))
```

```python
confusion_matrix(y_test, pred)
```

### Support Vector Classification

```python
from sklearn import svm
rfc = svm.SVC(kernel='linear', C = 1.0)
rfc.fit(X_train, y_train)
print('Ошибка Support Vector Classification составляет', (1 - rfc.score(X_test, y_test)) * 100, '%')
```

### Сравним результаты с другим классификатором Decision Tree Classifier

```python
from sklearn import tree
errors = []
for i in range(3, 40, 3):
    rfc = tree.DecisionTreeClassifier(max_depth = i)
    rfc.fit(X_train, y_train)
    print('Ошибка Support Vector Classification составляет', (1 - rfc.score(X_test, y_test)) * 100, '%')
    errors.append((1 - rfc.score(X_test, y_test))*100)
```

```python
min(errors) 
```

### Вывод:
* knn - ошибка около 6.4%
* Метод опорных векторов - ошибка 3%
